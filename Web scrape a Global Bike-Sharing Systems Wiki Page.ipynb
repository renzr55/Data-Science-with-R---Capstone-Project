````r
url <- "https://en.wikipedia.org/wiki/List_of_bicycle-sharing_systems"
# Get the root HTML node by calling the `read_html()` method with URL
url <- "https://en.wikipedia.org/wiki/List_of_bicycle-sharing_systems"
root_node <- read_html(url)

# Extract all child <table> nodes under the root node
table_nodes <- html_nodes(root_node, "table")

# Optional: Print each table to identify the correct one (if needed)
for (i in seq_along(table_nodes)) {
  print(html_table(table_nodes[[i]], fill = TRUE))
}

# Extract the first table, which contains the bike-sharing data
bike_sharing_table <- html_table(table_nodes[[1]], fill = TRUE)

# View the first few rows of the table to confirm
head(bike_sharing_table)

# Save the table as a CSV file for future use
write.csv(bike_sharing_table, "raw_bike_sharing_systems.csv", row.names = FALSE)

# Confirmation message
cat("Bike-sharing systems data has been saved to 'raw_bike_sharing_systems.csv'\n")
````

````r
Note that this HTML page at least contains three child `<table>` nodes under the root HTML node. So, you will need to use `html_nodes(root_node, "table")` function to get all its child `<table>` nodes:

```
<html>
  <table>(table1)</table>
  <table>(table2)</table>
  <table>(table3)</table>
  ...
</html>
```
````
